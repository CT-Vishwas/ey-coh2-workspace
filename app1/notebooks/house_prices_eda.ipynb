```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'C:\Users\VishwasKSingh\Workspace\ey-coh2-workspace\app1\data\house_prices_india.csv')
```

```python
# Convert number of bathrooms to integer
data['number of bathrooms'] = data['number of bathrooms'].fillna(0).astype(int)
```

```python
# How many unique postal codes are there in the dataset?
data['Postal Code'].nunique()
```

```python
# What is the average price of houses in the dataset grouped by postal code?
data.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head(10)
```

```python
# maxprice by postal code
data.groupby('Postal Code')['Price'].max().sort_values(ascending=False).head(10)
```

```python
# In the dataset, piechart of the distribution of the number of bedrooms
bedroom_counts = data['number of bedrooms'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(bedroom_counts, labels=bedroom_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Number of Bedrooms')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
# plot of Number of Schools Nearby vs postal code 5 hightest postal code
top_postal_codes = data.groupby('Postal Code')['Number of schools nearby'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_postal_codes.index, y=top_postal_codes.values)
plt.title('Average Number of Schools Nearby by Postal Code (Top 5)')
plt.xlabel('Postal Code')
plt.ylabel('Average Number of Schools Nearby')
plt.xticks(rotation=45)
plt.show()
```
