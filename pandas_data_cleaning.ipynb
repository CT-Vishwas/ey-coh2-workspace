```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a sample Series from a dictionary
data = {'A': 10, 'B': 20, 'C': 30, 'D':78}
series = pd.Series(data)
print(series)
```

```python
# Create a sample dataframe
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25,None, 30, "35"],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)
```

Exploratory Data Analysis & Cleaning

```python
df.info()
```

```python
# Convert Age Column data to integers
df['Age'] = pd.to_numeric(df['Age'], errors='coerce').astype('Int64')
```

```python
df.info()
```

```python
# Create a new Dataframe and fill null values in age column with 0
df_new = df.copy()
df_new['Age'] = df_new['Age'].fillna(0)
```
